```python
import cv2
import os
import time
from pathlib import Path
from PIL import Image
from facenet_pytorch import MTCNN
from torchvision.transforms import ToPILImage 

# Set name and number of images
person_name = input("Enter the name of the person: ")
num_images = int(input("How many images to capture? "))

# Set up paths
save_path = Path(f'/Users/rajataggarwal/Desktop/dataset/{person_name}')  
save_path.mkdir(parents=True, exist_ok=True)

# Initialize webcam and face detector
cap = cv2.VideoCapture(0)
mtcnn = MTCNN(keep_all=False)
to_pil = ToPILImage() 

print(f"Capturing {num_images} face images for {person_name}. Press 'x' to stop early.")

count = 0
last_capture_time = 0
interval = 0.5  # 0.5 seconds between captures (buffer)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    face = mtcnn(Image.fromarray(img_rgb))

    # Show webcam feed
    cv2.imshow("Face Capture", frame)

    # If a face is detected and time interval passed, save it
    current_time = time.time()
    if face is not None and current_time - last_capture_time >= interval:
        face_crop = to_pil(face.squeeze(0))  # Convert tensor to PIL image
        face_crop.save(save_path / f"{count + 1}.jpg")
        print(f"Saved image {count + 1}")
        count += 1
        last_capture_time = current_time

    # Break if done or user presses 'x'
    if count >= num_images or cv2.waitKey(1) & 0xFF == ord('x'):
        break

cap.release()
cv2.destroyAllWindows()
print(f"Saved {count} face images for {person_name} in '{save_path}'.")
```
